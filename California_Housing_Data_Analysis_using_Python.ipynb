##  California Housing Data Analysis

### Objective
The goal of this project is to analyze the California housing dataset and identify factors that influence median house prices.

### Dataset
- Source: California Housing Dataset
- Rows: ~17,000
- Features: Location, population, housing age, income, and house value

### Key Insights
- Median house values are right-skewed, meaning most houses are priced lower with fewer expensive houses.
- Population size does not show a strong linear relationship with house prices.
- Housing age has some influence, but not a dominant factor.
- Location-based features (latitude) play a significant role in housing prices.

### Conclusion
This analysis demonstrates how exploratory data analysis (EDA) can reveal meaningful patterns in real-world data using Python. The project uses basic yet powerful techniques suitable for entry-level data analyst roles.
